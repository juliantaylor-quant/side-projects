# Getting Started

```python
#Import necessary libraries
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

#set tickers variable - Apple, Microsoft, Google, Amazon, Tesla
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']

#set start and end dates for data retrieval
end_date = datetime.now() 
start_date = end_date - timedelta(days=2*365)
print(end_date)
print(start_date)

#Create empty DataFrame to store close prices
close_prices = pd.DataFrame()

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    close_prices[ticker] = data['Close']

close_prices.head()

#Save the close prices to a CSV file
close_prices.to_csv('close_prices.csv', index=True)
```

# Close dates EDA

```python
#Load the close prices from the CSV file
close_prices = pd.read_csv('close_prices.csv', index_col=0, parse_dates=True)
close_prices.head()
```

```python
#Find summary statistics for the close prices of each stock
close_summary = pd.DataFrame(index=tickers, columns=['Min', 'Max', 'Average', 'Median'])

for i in range(len(tickers)):
    close_summary.loc[tickers[i], 'Min'] = close_prices[tickers[i]].min()
    close_summary.loc[tickers[i], 'Max'] = close_prices[tickers[i]].max()
    close_summary.loc[tickers[i], 'Average'] = close_prices[tickers[i]].mean()
    close_summary.loc[tickers[i], 'Median'] = close_prices[tickers[i]].median()

print(close_summary)

# Microsoft had the highest average close price, while Amazon had the lowest average close price. 
# Microsoft also had the highest maximum close price, while Amazon had the lowest minimum close price.
```

```python
#Find the daily returns for each stock
daily_returns = close_prices.pct_change().dropna()
daily_returns.head()
```
